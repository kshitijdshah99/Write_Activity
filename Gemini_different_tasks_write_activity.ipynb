```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from crewai import Agent, Task, Crew, Process
from langchain.tools import DuckDuckGoSearchRun

# Set gemini pro as llm
llm = ChatGoogleGenerativeAI(model="gemini-pro",
                             verbose=True,
                             temperature=0.5,
                             google_api_key="AIzaSyC1v-IIhbBPlFzJnR6QKSk54mU6Cm_XsPM")

# Create searches
tool_search = DuckDuckGoSearchRun()

# Define Agents
essay_writer = Agent(
    role='Creative Essay Writer for Children',
    goal='Generate engaging essays and stories for children',
    backstory='Experienced in crafting imaginative and child-friendly content',
    verbose=True,
    allow_delegation=True,
    llm=llm,
    tools=[tool_search]
)

story_creator = Agent(
    role='Story Creator for Children',
    goal='Create captivating stories for children to enjoy',
    backstory='Skilled in developing plots and characters that resonate with young readers',
    verbose=True,
    allow_delegation=True,
    llm=llm,
    tools=[tool_search]
)

creative_thinker = Agent(
    role='Creative Thinker for Children',
    goal='Generate innovative and imaginative ideas for various purposes',
    backstory='Nurtured creativity through brainstorming and exploration',
    verbose=True,
    allow_delegation=True,
    llm=llm,
    tools=[tool_search]
)

kids_input = input("Enter your query: ")

# Define Tasks
essay_writing_task = Task(
    description='Write an essay on the importance of imagination in children',
    agent=essay_writer,
    expected_output='A well-written essay highlighting the significance of imagination in child development'
)

story_creation_task = Task(
    description='Create a short story about a magical adventure in a faraway land',
    agent=story_creator,
    expected_output='Tell me A captivating story featuring magical elements and exciting adventures'
)

idea_generation_task = Task(
    description='Generate creative ideas for a children\'s art project',
    agent=creative_thinker,
    expected_output='A list of imaginative ideas suitable for a children\'s art project'
)

# Create a Crew
children_creative_crew = Crew(
    agents=[essay_writer, story_creator, creative_thinker],
    tasks=[essay_writing_task, story_creation_task, idea_generation_task],
    verbose=True,
    process=Process.sequential  # You can define the process if needed, e.g., sequential, parallel, etc.
)

print("Crew: Working on query generation Task")
code_output = children_creative_crew.kickoff()
```
